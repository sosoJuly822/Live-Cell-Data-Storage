```python
import pickle

def load_dict_pickle(file_path):
    with open(file_path, 'rb') as f:
        return pickle.load(f)
```

```python
pkl_file = '/home/liuycomputing/wsp_sequencing/codes/process_20251218/crispr_select/crispr_select_before_count_reads_for_perfect_recovery.pkl'

target_ids = [1, 3340, 6800, 10200, 13600, 17000, 20400, 23800, 27200, 30600]  # 目标ID列表，默认为[1]
```

```python
result = load_dict_pickle(file_path=pkl_file)
print(result['whole_finded_ids'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

IDs = target_ids

data  = [result['whole_finded_ids'][i-1] for i in IDs]

x = np.arange(len(IDs)) + 1

plt.figure(figsize=(10, 4))

# Before
plt.boxplot(
    data,
    widths=0.6,
    showfliers=True,
    patch_artist=True,
    boxprops=dict(facecolor="white"),
    medianprops=dict(color="orange")
)


plt.xticks(x, IDs)

plt.xlabel("IDs")
plt.ylabel("Reads")
plt.yscale("log")   # 数据跨度大，强烈推荐
plt.tight_layout()
```
